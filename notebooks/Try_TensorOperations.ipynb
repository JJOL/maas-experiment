```python
f = torch.rand(4, 5, 8)
f = f.to(device)
f
```

```python
f[:, 3, :]
```

```python
mask = torch.zeros(4, 5, 8)
mask[0, 4, :] = torch.ones(1)
mask[1, 4, :] = torch.ones(1)
mask[2, 2, :] = torch.ones(1)
mask[3, 3, :] = torch.ones(1)
mask = mask.to(device)
mask
```

```python
f_masked = f * mask
f_masked
```

```python
output = torch.sum(f_masked, dim=1)
output
```

```python
output.size()
```
